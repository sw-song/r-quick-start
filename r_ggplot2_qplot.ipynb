## 데이터 시각화 - qplot

ggplot2 라이브러리를 볼러옵니다.

```r
library(ggplot2)
```

`ggplot2`에서 제공하는 `mpg` 데이터셋을 사용합니다.

```r
head(mpg)
```

`qplot`은 `quick plot`의 줄임 표현이며, 말 그대로 빠르게 시각화 할 수 있도록 가벼운 문법을 제공합니다.

우선, 몇 가지 변수(feature)들의 빈도를 표시해보겠습니다.

```r
qplot(data=mpg, x=manufacturer)
```

```r
qplot(data=mpg, x=fl)
```

```r
qplot(data=mpg, x=year)
```

좀 이상합니다. 제조년도가 양 극단에 몰려 있는 것을 볼 수 있습니다.

데이터로 확인해봅니다.

```r
table(mpg$year)
```

1999년, 2008년식 차량 데이터만 포함되어 있습니다. 이 점은 분석할 때에 유의할 점이 되겠습니다.

```r
qplot(data=mpg, x=class)
```

이렇게 x축에 표시할 데이터만 입력해주면 데이터의 빈도를 표시해줍니다.

이번에는 y 값도 함께 넣어보겠습니다. 

```r
qplot(data=mpg, x=class, y=hwy)
```

`class`는 사이즈, 형태와 같은 차종 데이터를 담고 있습니다.

```r
table(mpg$class)
```

위 시각화에서는 차종별로 고속도로 연비를 확인할 수 있습니다.

이를 산점도라 하며 예시로 `pickup` 트럭은 연비가 다른 차종에 비해 떨어지는 것을 보여줍니다.

그런데 산점도보다 데이터 분포를 효과적으로 확인할 수 있는 다른 시각화 방식이 있습니다.


```r
qplot(data=mpg, x=class, y=hwy, geom='boxplot', color=class)
```

`boxplot`입니다. 각 차종(class)별 연비의 평균과 이상치(점 표시)를 효과적으로 시각화하고 있습니다.

시각화 방식은 매우 중요합니다. 변수간 평균치를 확인해야 할 때 아래와 같이 선으로 표현해버리면 데이터 분포를 알 수 없습니다.

```r
qplot(data=mpg, x=class, y=hwy, geom='line', color=class)
```
